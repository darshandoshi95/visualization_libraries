```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pa
filename='/data/Iris_Data.csv'
data=pa.read_csv(filename) 
```

```python
#scatter plot created from pandas series
plt.plot(data.sepal_length,data.sepal_width,ls='',marker='o')
```

```python
#we can add multiple layers of data to same scatter plot
plt.plot(data.sepal_length,data.sepal_width,ls='',marker='*',label='sepal')
plt.plot(data.petal_length,data.petal_width,ls='',marker='o',label='petal')
```

```python
#histogram plot
plt.hist(data.sepal_length,bins=25)
```

```python
#every feature of matplotlib can be customized
fig, ax=plt.subplots()
ax.barh(np.arange(10),data.sepal_width.iloc[:10])
ax.set_yticks(np.arange(0.4,10.4,1.0))
ax.set_yticklabels(np.arange(1,11))
ax.set(xlabel='xlabel', ylabel='ylabel',title='title')
```

```python
#incorporsting statistical calculations
(data.groupby('species').mean().plot(color=['red','black','blue','green'],fontsize=10.0,figsize=(4,4)))
```

```python
#statistical plotting with seaborn
sns.jointplot(x='sepal_length',y='sepal_width',data=data)
```

```python
sns.pairplot(data,hue='species',size=3)
```
